```python
import openpyxl

original_file = "bu-dataset.xlsx"

wb = openpyxl.load_workbook(original_file)
ws = wb['Sheet1']
```

```python
import re
column = ws['K']
links = ws['R']
for cell, l in zip(column, links):
    link = re.search("=HYPERLINK\(\"(.*)\",", cell.value) # get the link
    try:
        l.value = link.group(1) # write last column
    except:
        pass
#     print(ws.cell(cell.row, max_column))

links[0].value = "vid_links"


wb.save("links-extracted.xlsx")
```

```python
import numpy as np
import pandas as pd

source = "links-extracted.xlsx"
df = pd.read_excel(source).iloc[:, 1:]
df.describe()
```

```python
df = df[(df.Consultant != "------------") & (df.Consultant != "============")]
df = df.drop(["Combined","Separate","Combined.1","Separate.1"], 1)
df.head()
# 9,10,15,16
df.to_excel("clean-dataset.xlsx")
```
